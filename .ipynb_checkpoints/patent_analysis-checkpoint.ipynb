```python
etree=et.parse(r"C:\Users\Lenovo\Documents\GitHub\PatentAnalysis\ipg221011.xml")
```

```python
root=etree.getroot()
```

```python
[elem.tag for elem in root.iter()]
```

```python
list_of_inventions=[]
for item in root.iter('us-bibliographic-data-grant'):
    for x in item.iter('invention-title'):
        list_of_inventions.append({'Title':x.text})
```

```python
series=pd.DataFrame(data=list_of_inventions)
series.head()
```

```python
patent_and_claims=[]
patent_no=0
for item in root.iter('claims'):
    list_of_claims=[]
    for x in item.iter('claim'):
        for y in x:
            list_of_claims.append(y)
    patent_and_claims.append({'patent_no':patent_no,'claims':list_of_claims})
print(len(patent_and_claims))
```
