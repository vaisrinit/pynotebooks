```python
from sympy import *
from sympy.plotting import plot3d
```

```python
x = symbols('x')
f = 2*x + 1
plot(f)
```

```python
x = symbols('x')
f = x**2 + 1
plot(f)
```

```python
x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

```python
summation = sum(2*i for i in range(1,6))
print(summation)
```

```python
x = symbols('x')
expr = x**2 / x**5
print(expr)
```

```python
x = symbols('x')
f = 1 / x
result = limit(f, x, oo)
print(result) # 0
```
